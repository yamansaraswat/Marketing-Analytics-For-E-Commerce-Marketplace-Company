##### 1 Understand the trends/seasonality of sales, quantity by category, location, month, week, day, time, channel, payment method etc…

```python
order_order_item= pd.merge(order_item , order , on= 'order_id', how= 'left')
order_item_prod= pd.merge(order_order_item, product , on = 'product_id' , how = 'left')
```

```python
order_item_prod.head(2)
```

```python
#Quantity by category
qty_by_cat = order_item_prod.groupby('product_category_name').agg({'product_id': 'count' , 'price':'sum'}).reset_index().rename(columns={'product_id':'quantity','price':'sales'})
qty_by_cat
```

```python
order_payment.head(2)
```

```python
#payment method
qty_by_paymentmethod = order_payment.groupby('payment_type').agg({'payment_value':'sum','order_id' : 'count' }).reset_index().rename(columns={'order_id':'quantity','payment_value':'sales'})
qty_by_paymentmethod
```

```python
#quantity by time
qty_by_tp = order_item_prod.groupby('year_month').agg({'product_id': 'count' , 'price':'sum'}).reset_index().rename(columns={'product_id':'quantity','price':'sales'})
qty_by_tp.head(2)
```

#### 2a. Divide the customers into groups based on the revenue generated 

```python
cust_group = order_item_prod.loc[:,['customer_id','final price']].rename(columns={'final price':'total_spent_amount'})
cust_group
```

```python
cust_group['cust_rev_gtd'] = pd.cut(cust_group.total_spent_amount,
                               bins= np.arange(0 , 22000, 2000),
                               labels=['0-2000','2000-4000','4000-6000','6000-8000',
                                       '8000-10000','10000-12000','12000-14000','14000-16000','16000-18000',
                                      '18000-20000'])
cust_group
```

###### 5a. Which categories (top 10) are maximum rated & minimum rated?

```python
order_item_prod_rating=pd.merge(order_item_prod, order_review ,how= 'left', on = 'order_id')
cat_review = order_item_prod_rating.groupby('product_category_name')['review_score'].sum().reset_index()
max_rating = cat_review.sort_values(by='review_score',ascending=False).head(10).reset_index()
max_rating
```

```python
min_rating = cat_review.sort_values(by='review_score',ascending=True).head(10).reset_index()
min_rating
```

##### 5.b Which products (top10) are maximum rated & minimum rated?

```python
product_review = order_item_prod_rating.groupby('product_id')['review_score'].sum().reset_index()
maximum_rating = product_review.sort_values(by='review_score',ascending=False).head(10).reset_index()
maximum_rating
```

```python
min_rating = product_review.sort_values(by='review_score',ascending=True).head(10).reset_index()
min_rating
```
